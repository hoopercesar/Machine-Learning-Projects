## Cargando el Modelo

```python
# importando joblib. 
import joblib 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargando el modelo
modelo = joblib.load('modelo_enero2022.pkl')
```

### Generando el camino para traer los archivos, via comandos
Estos pasos nos permiten traer los archivos xlsx desde donde sea que estén. 

```python
# OJO: con el back-slash:
# path ='C:\Users\Cesar Hooper\Desktop\github\proyecto_MARCELO/'

# cambiar backslash para slash
path ='C:/Users/Cesar Hooper/Desktop/.../proyecto_MARCELO/'
```

```python
# cargando datos los datos de febrero 2022.
datos = pd.read_excel(path + 'DBMLJAN22.xlsx')
```

```python
datos.head(10)
```

```python
# comandos para quitar fecha, hora y eliminar NaN
dataset = datos.dropna().copy()
fecha = dataset['Unnamed: 0']
hora = dataset['Unnamed: 1']
target = dataset['Unnamed: 16'] # cargué datos con clasificación
dataset = dataset.drop(['Unnamed: 0', 'Unnamed: 1', 'Unnamed: 16'], axis=1)
```

```python
# con el comando shape comprobamos que los datos tienen el formato de 14 columnas y la misma cantidad de filas
dataset.shape, target.shape, fecha.shape, hora.shape
```

```python
# randomizamos los datos
dataset_randomized = dataset.sample(frac=1)
dataset_randomized.head()
```

```python
# importando librería para normalización de datos
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

#  (fiteamos) los datos randomizados
scaler.fit(dataset_randomized)
```

```python
# normalización de los datos no randomizados
dataset_normalized = scaler.transform(dataset)
dataset_normalized
```

```python
# aplicación modelo sobre los datos limpios y normalizados.
# recordar que el modelo se cargó con el nombre de modelo

# este comando entrega el resultado de la clasificación -1, 0 ó 1
y_pred = modelo.predict(dataset_normalized)

# este comando entrega la probabilidad de cada resultado
y_proba = modelo.predict_proba(dataset_normalized)
```

```python
# verificamos dimensiones de los resultados
# lo importante es que tengan misma cantidad de filas.
y_pred.shape, y_proba.shape
```

```python
# estos son los resultados de la clasificación
y_pred
```

```python
# resultado de las probabilidades
y_proba
```

## Resultados
Para poder entender mejor los resultados, vamos a crear una tabla de resultados

```python
# resultados
resultados = pd.DataFrame()
resultados['fecha'] = fecha
resultados['hora'] = hora
resultados['Prediccion'] = y_pred
resultados['%pred: -1'] = y_proba[:,0]
resultados['%pred: 0'] = y_proba[:,1]
resultados['%pred: 1'] = y_proba[:,2]

resultados.head(10)
```

```python
resultados.to_excel('RESULTADOS_ENERO2022.xlsx', index=False)
```
